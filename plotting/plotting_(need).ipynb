## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values') # define x and y variables and aesthetic stuff 
plt.xlabel('x-values') # give x-axis a name of x-values
plt.ylabel('y-values') # give y-axis a name of y-values
plt.title('X vs Y') # giving the graph a title
plt.grid(alpha=.25) # grid at the back and the transparency of the grid lines
plt.legend(loc='upper left') # insert legend and where it is located 
plt.show() # did not change anything 
```

### 1.2 Adding another plot

```python
plt.plot(x, y_2, color='blue', label='Y2 values') # plots it separately 
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')                          # inserted here 
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75                 # must define the error before plotting 


plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,      # insert error bar function here. note you are plotting y3 with error
             color='green', label='Y3 with errors')   # not each line's error 
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('X graphs')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.7 We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

plt.twinx()                                         # This creates a new y-axis for the plot that comes after
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

```python
x = np.linspace(0, 6, 100) # change x values to this range instead. linspace returns values with evenly spaced numbers over a
x2 = x**2 # specified interval. syntax is (start,stop, no. of samples)
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

plt.twinx()                                         # This creates a new y-axis for the plot that comes after
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

### 1.8 Saving to disc

```python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0] # select all rows first column 
plt.style.use('fivethirtyeight')
anomaly = data[:, -1] # all rows last column 
plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```
